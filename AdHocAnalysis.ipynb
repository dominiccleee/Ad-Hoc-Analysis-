```python
import pandas as pd
import numpy as np
import json
```

```python
# main dataframe
f = open('transaction-data-adhoc-analysis.json')
data = json.loads(f.read())
df = pd.DataFrame.from_dict(pd.json_normalize(data), orient='columns')
f.close()

df
```

```python
# aid to look for unit price
df2 = df
df2.transaction_items = df2.transaction_items.astype(str)
df2[df2['transaction_items'].str.contains('Yummy')]
```

```python
# Item Unit Prices (manually computed)
price_df = pd.DataFrame({'Product': ['Candy City Gummy Worms', 'Candy City Orange Beans', 'Exotic Extras Beef Chicharon', 'Exotic Extras Kimchi and Seaweed', 
                                     'HealthyKid 3+ Gummy Vitamins', 'HealthyKid 3+ Nutrional Milk', 'HealthyKid 3+ Yummy Vegetables'],
                        'Price': [150, 199, 1299, 799, 1500, 1990, 500] })
price_df
```

```python
#item dataframe
item_df = df[['transaction_items', 'transaction_value', 'transaction_date']].copy()

#transaction items
item_df = item_df.drop('transaction_items', axis=1).join(df.transaction_items.str.split(";",expand=True).stack().reset_index(drop=True,level=1).rename('transaction_item'))
item_df["quantity"] = item_df.transaction_item.str.extract(r'(\d+)', expand=True)
item_df['transaction_item'] = item_df.transaction_item.str.replace(r'(x\d+)'," ", regex=True)
item_df['transaction_item'] = item_df.transaction_item.str.replace(","," ").str.replace("("," ").str.replace(")"," ")
item_df.quantity = item_df.quantity.astype(int)

# transaction date
item_df['year']= item_df.transaction_date.str.split('/', expand=True)[0]
item_df['month']= item_df.transaction_date.str.split('/', expand=True)[1] 
item_df['day']= item_df.transaction_date.str.split('/', expand=True)[2] 

item_df
```

```python
january = item_df.loc[item_df['month'] == "01"].groupby('transaction_item')["quantity"].sum()
january
```

```python
item_sold_per_month = item_df.groupby(['month','transaction_item'])['quantity'].agg('sum').reset_index()
item_sold_per_month
```

### Quantities of Item Sold Per Month Table

```python
# Quantities of Item Sold Per Month Table
output = pd.pivot_table(data=item_sold_per_month, 
                        index=['transaction_item'], 
                        columns=['month'], 
                        values='quantity')
output
```

```python
price_df.loc[price_df['Product'].str.contains("Worm")]["Price"]
```

```python
def computeSales(prod, q):

    if "Worm" in prod:
      sales = price_df.loc[price_df['Product'].str.contains("Worm")]["Price"] * q
    elif "Orange" in prod:
      sales = price_df.loc[price_df['Product'].str.contains("Orange")]["Price"] * q
    elif "Beef" in prod:
      sales = price_df.loc[price_df['Product'].str.contains("Beef")]["Price"] * q
    elif "Kimchi" in prod:
      sales = price_df.loc[price_df['Product'].str.contains("Kimchi")]["Price"] * q
    elif "Vitamins" in prod:
      sales = price_df.loc[price_df['Product'].str.contains("Vitamins")]["Price"] * q
    elif "Milk" in prod:
      sales = price_df.loc[price_df['Product'].str.contains("Milk")]["Price"] * q
    elif "Vegetables" in prod:
      sales = price_df.loc[price_df['Product'].str.contains("Vegetables")]["Price"] * q
   
    return sales
```

```python
item_sales_per_month = item_sold_per_month
item_sales = item_sales_per_month.apply(lambda x: computeSales(x.transaction_item, x.quantity), axis=1)
item_sales = item_sales.replace(np.nan,0)
item_sales['Total Sales']= item_sales.sum(axis=1)
item_sales_per_month['Total Sales'] = item_sales['Total Sales']
item_sales_per_month
```

### Item Sales Per Month Table

```python
# Item Sales Per Month Table
output = pd.pivot_table(data=item_sales_per_month, 
                        index=['transaction_item'], 
                        columns=['month'], 
                        values='Total Sales')
output
```

```python
#customer dataframe
customer_df = df[['name', 'transaction_date']].copy()

# transaction date
customer_df ['year']= customer_df.transaction_date.str.split('/', expand=True)[0]
customer_df ['month']= customer_df.transaction_date.str.split('/', expand=True)[1] 
customer_df ['day']= customer_df.transaction_date.str.split('/', expand=True)[2] 

customer_df 
```

```python
customer_df = customer_df.drop(["transaction_date","year","day"], axis=1)
customer_df = customer_df.drop_duplicates()
customer_df
```

```python
jan_df = customer_df.loc[customer_df["month"] == "01"]
feb_df = customer_df.loc[customer_df["month"] == "02"]
mar_df = customer_df.loc[customer_df["month"] == "03"]
apr_df = customer_df.loc[customer_df["month"] == "04"]
may_df = customer_df.loc[customer_df["month"] == "05"]
jun_df = customer_df.loc[customer_df["month"] == "06"]
```

```python
#Repeater Count
jan_rep = 0
feb_rep = pd.merge(jan_df["name"],feb_df["name"], how='inner').size
mar_rep = pd.merge(feb_df["name"],mar_df["name"], how='inner').size
apr_rep = pd.merge(mar_df["name"],apr_df["name"], how='inner').size
may_rep = pd.merge(apr_df["name"],may_df["name"], how='inner').size
jun_rep = pd.merge(may_df["name"],jun_df["name"], how='inner').size

print(jan_rep, feb_rep, mar_rep, apr_rep, may_rep, jun_rep)
```

```python
jf_df = pd.concat([jan_df,feb_df])
jf_df = jf_df.drop(["month"], axis=1).drop_duplicates()
jfm_df = pd.concat([jan_df,feb_df,mar_df])
jfm_df = jfm_df.drop(["month"], axis=1).drop_duplicates()
jfma_df = pd.concat([jan_df,feb_df,mar_df,apr_df])
jfma_df = jfma_df.drop(["month"], axis=1).drop_duplicates()
jfmam_df = pd.concat([jan_df,feb_df,mar_df,may_df])
jfmam_df = jfmam_df.drop(["month"], axis=1).drop_duplicates()
```

```python
#Inactive Count
jan_ina = 0
feb_ina = pd.merge(jan_df["name"], feb_df["name"], how = 'outer' ,indicator=True).loc[lambda x : x['_merge']=='left_only']["name"].size
mar_ina = pd.merge(jf_df["name"], mar_df["name"], how = 'outer' ,indicator=True).loc[lambda x : x['_merge']=='left_only']["name"].size
apr_ina = pd.merge(jfm_df["name"], apr_df["name"], how = 'outer' ,indicator=True).loc[lambda x : x['_merge']=='left_only']["name"].size
may_ina = pd.merge(jfma_df["name"], may_df["name"], how = 'outer' ,indicator=True).loc[lambda x : x['_merge']=='left_only']["name"].size
jun_ina = pd.merge(jfmam_df["name"], jun_df["name"], how = 'outer' ,indicator=True).loc[lambda x : x['_merge']=='left_only']["name"].size

print(jan_ina, feb_ina, mar_ina, apr_ina, may_ina, jun_ina)
```

```python
#Engaged Count
jan_eng = jan_df["name"].size
feb_eng = pd.merge(feb_df["name"], jan_df["name"], how='inner').size
mar_eng = pd.merge(mar_df["name"], jf_df["name"], how='inner').size
apr_eng = pd.merge(apr_df["name"], jfm_df["name"], how='inner').size
may_eng = pd.merge(may_df["name"], jfma_df["name"], how='inner').size
jun_eng = pd.merge(jun_df["name"], jfmam_df["name"], how='inner').size

print(jan_eng, feb_eng, mar_eng, apr_eng, may_eng, jun_eng)
```

```python
transactions_df = pd.DataFrame(
    {
        "Jan": [jan_rep, jan_ina, jan_eng],
        "Feb": [feb_rep, feb_ina, feb_eng],
        "Mar": [mar_rep, mar_ina, mar_eng],
        "Apr": [apr_rep, apr_ina, apr_eng],
        "May": [may_rep, may_ina, may_eng],
        "Jun": [jun_rep, jun_ina, jun_eng],
    },
    index=["Repeater", "Inactive", "Engaged"],
)
```
